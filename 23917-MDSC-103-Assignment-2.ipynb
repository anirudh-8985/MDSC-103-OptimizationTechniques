# Performing Gradient Descent Convergence Function on single variable functions and plotting the X*'s on  the graphs




Function: $y = x^2$   on   [-5,5]
    

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,100)
f = lambda x: x**2     #Defining lambda function for x**2
plt.plot(x,f(x))
plt.grid()
plt.title("Convergence Function")
plt.xlabel("x values")
plt.ylabel("Function Values")

x0 = 4                              # initial point
plt.scatter(x0,f(x0), color = 'g')  # Plotting initial point on the graph

gd = lambda x: x -(0.1)*(2*x)       #Lambda function for the function x**2
i = 1

while (i <= 100):                 # iterating 100 times for calculating X*
    x1 = gd(x0)
    if((x0-x1) < 1e-5):
        print(x1)
        break
    plt.scatter(x1,f(x1), color = 'r')
    x0 = x1
    i += 1
```

                                          GD_FUNCTION
GD_function(function,def_function,initial_point,lr,tolerance,no_of_iteration) is function for plotting and calculating the Gradient decent of one variable function
 
    It will take 6 parameters:
    1) function         = Is the function for which the Gradient Function is to be implemented.
    2) def_Function     = Gradient of the function
    3) inital_point     = Initial Point to be started
    4) lr               = eta value
    5) Tolerance        = Difference between xk and x(k-1)
    6) No.of iterations = No.of iteration the xk is calculated

-> If the X* is not converges in the given iteration, then the output will be "Out of iteration!" and it will display the last Xk value.

```python
# GD_function(function,def_function,initial_point,lr,tolerance,no_of_iteration) is function for plotting and calculating the Gradient decent of one variable function

def GD_funct(func,df, init_p, lr , tol, n ):   
    x = np.linspace(-5,5,100)
    plt.plot(x,func(x))
    plt.grid()
    plt.title("Gradient Decent Convergence Function")
    plt.xlabel("x values")
    plt.ylabel("Function Values")
    
    gd = lambda x0,lr: x0 - (lr*df(x0))     # Gradent descent function
    i = 1
    plt.scatter(init_p,func(init_p), color = 'g')
    while (i<n):
        x1 = gd(init_p,lr)
        plt.scatter(x1,func(x1), color = 'r')
        if(np.abs(init_p-x1) < 1e-5):          
            print("The value of x* is :",x1)
            break
        init_p = x1
        i += 1
        if(i == n-1):
            print("Out of iteration! The last x* value is :",x1)
        
        
    
```

 Function : $ 2x^3 - 3x $

```python
f = lambda x :  2*(x**2)-3*x
df = lambda x : 4*x - 3

GD_funct(f,df,4,0.1,1e-5,100) # GD_function(funct,def_func,init_point,lr,tolerance,no_of_iter)

  
```

Function : $x^3$

```python
f = lambda x :  x**3
df = lambda x : 3*x**2

GD_funct(f,df,4,0.001,1e-7,10000)   # GD_function(funct,def_func,init_point,lr,tolerance,no_of_iter)

  
```
