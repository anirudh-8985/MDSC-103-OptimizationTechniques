Minimize $f(x)= x1^3 + x2^3$ - 3x1x2

Subject to $$3x1 + x2 \le3, \\
            5x1 - 3x2 \le5, \\
            x1,x2\ge0$$

```python
# Minimizing the function with constraints

import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import minimize

f = lambda x: x[0]**3 + x[1]**3 - 2*x[0]*x[1]
cons = ({'type': 'ineq', 'fun' : lambda x: -3*x[0]-x[1]+3},
       {'type': 'ineq', 'fun' : lambda x: -5*x[0]+3*x[1]+5})

bnds = ((0,None),(0,None))
X0 = (1,0)

res = minimize(f,X0, bounds=bnds, constraints = cons)
res
                  
```

```python
# Minimizing the function without constraints

res2 = minimize(f,X0)
res2
```
