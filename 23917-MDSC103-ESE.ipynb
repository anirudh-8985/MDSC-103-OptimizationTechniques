Given function:
$$f(x1,x2) = 4x1 + 6x2 - 2x1^2 - 2x1x2 - 2x2^2$$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from mpl_toolkits.mplot3d import Axes3D


f = lambda x1,x2: 4*x1 + 6*x2 - 2*(x1**2) - 2*x1*x2 - 2*(x2**2) # Function declaration
x1 = np.linspace(-500,10000,1000)
x2 = np.linspace(-500,10000,1000)
X, Y = np.meshgrid(x1,x2)
F = f(X,Y)

fig = plt.figure(figsize=(15,15))
ax = plt.subplot(1,2,1, projection='3d') # set up the axis as a 3D plot

ax.plot_surface(X,Y,F, cmap='gist_rainbow')    # Plotting 3D plot


ax.set_xlabel('x1', labelpad=10)
ax.set_ylabel('x2', labelpad=30)
ax.set_zlabel('f(X)', labelpad=10)
plt.title("3D plot of the given function")

plt.show()
```

```python
# converting the given function to minimization and Plotting  the Graph

f = lambda x1,x2: -(4*x1 + 6*x2 - 2*(x1**2) - 2*x1*x2 - 2*(x2**2))   # Minimization function
x1 = np.linspace(0,10000,1000)
x2 = np.linspace(0,10000,1000)
X, Y = np.meshgrid(x1,x2)
F = f(X,Y)

fig = plt.figure(figsize=(10,10))
ax = plt.subplot(1,2,1, projection='3d') # set up the axis as a 3D plot
ax.scatter3D(0,0,f(0,0), c="black",marker="*",s =100) 
ax.plot_surface(X,Y,F, cmap='gist_rainbow')  # Plotting 3D plot


ax.set_xlabel('x1', labelpad=10)
ax.set_ylabel('x2', labelpad=30)
ax.set_zlabel('f(X)', labelpad=10)
plt.title("3D plot of the Minimization function of f")
plt.show()
```

```python
# Gradient Descent Function 
f = lambda x1,x2: -(4*x1 + 6*x2 - 2*(x1**2) - 2*x1*x2 - 2*(x2**2))  #minimization function
df_x1 = lambda x1,x2 : -(4 - 4*x1 - 2*x2)                         # differntiation of f w.r.t x1
df_x2 = lambda x1,x2 : -(-6 - 2*x1 - 4*x2)                        # differntiation of f w.r.t x1
 
def Gradient_des(f,df_x1,df_x2,init_x1,init_x2,n,lr):   # Function declaration

    x1 = np.linspace(-5,5,100)
    x2 = np.linspace(-5,5,100)
    fig = plt.figure(figsize=(10,10))
    ax = plt.subplot(1,2,1, projection='3d')  # Setting up the axis
    ax.plot_surface(X,Y,F, cmap='gist_rainbow')  # Plotting the #D plot
    ax.set_xlabel('x1', labelpad=10)
    ax.set_ylabel('x2', labelpad=30)
    ax.set_zlabel('f(X)', labelpad=10)
    
    gd_x1 = lambda x0,x1,lr: x0 - (lr*df_x1(x0,x1))    # Gradient_dec function for x1
    gd_x2 = lambda x0,x1,lr: x0 - (lr*df_x2(x0,x1))    # Gradient_dec function for x1
    i = 1
    ax.scatter3D(init_x1,init_x2,f(init_x1,init_x2), c="black",marker="*",s =100)   # Plotting the marker at first initial points
    
    while(i<n):    # while loop for no_of iteration
        x1 = gd_x1(init_x1,init_x2,lr)    
        x2 = gd_x1(init_x1,init_x2,lr)
        ax.scatter3D(x1,x2,f(x1,x2), c="black",marker="*",s =100)    # Plotting the marker for the next iteration
        if(np.abs(init_x1-x1) < 1e-5 or np.abs(init_x2-x2) < 1e-5 ):     # if the values are less then lr then break     
            print("The value of x1* is :",x1)   # Printing x1 value
            print("The value of x2* is :",x2)    # Printing x2 value
            break
        init_x1 = x1                        # Changing the init point to the next value
        init_x2 = x2

        i += 1
        if(i == n-1):        # IF out of iteration then exit
            print("Out of iteration! The last x1* value is :",x1)   # Printing x1 value
            print("The value of x2* is :",x2)   # Printing x2 value
        
# Calling the function       
Gradient_des(f,df_x1,df_x2,6,5,1000,1e-5)  # here : 6,5 are initial points and 1000 = no.of iterations and 1e-5 is lr.
```
